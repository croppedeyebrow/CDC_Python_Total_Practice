## 파이썬의 산술 계산

### x + y   # 덧셈: x와 y를 더합니다.
### x - y   # 뺄셈: x에서 y를 뺍니다.
### x * y   # 곱셈: x와 y를 곱합니다.
### x / y   # 나눗셈: x를 y로 나눕니다. 결과는 항상 float입니다.
### x // y  # 정수 나눗셈: x를 y로 나누고 몫의 정수 부분만 반환합니다. 
### x % y   # 모듈로(나머지): x를 y로 나눈 나머지를 반환합니다.
### -x      # 단항 음수: x의 부호를 바꿉니다.
### +x      # 단항 양수: x의 값을 그대로 반환합니다. (주로 생략됨)
### abs(x)  # 절대값: x의 절대값을 반환합니다.
### divmod(x,y)  # 몫과 나머지: x를 y로 나눈 몫과 나머지를 튜플로 반환합니다.
### pow(x,y)  # 거듭제곱: x의 y승을 계산합니다. x**y와 같습니다.
### x ** y  # 거듭제곱: x의 y승을 계산합니다. pow(x,y)와 같습니다.

```python
# 예제 값 설정
x = 10
y = 3

print("덧셈 (+):")
print(x + y)  # 13

print("\n뺄셈 (-):")
print(x - y)  # 7

print("\n곱셈 (*):")
print(x * y)  # 30

print("\n나눗셈 (/):")
print(x / y)  # 3.3333333333333335 (float 결과)

print("\n정수 나눗셈 (//):")
print(x // y)  # 3 (몫의 정수 부분)

print("\n모듈로 (나머지) (%):")
print(x % y)  # 1

print("\n단항 음수 (-):")
print(-x)  # -10

print("\n단항 양수 (+):")
print(+x)  # 10 (변화 없음)

print("\n절대값 (abs()):")
print(abs(-x))  # 10

print("\n몫과 나머지 (divmod()):")
print(divmod(x, y))  # (3, 1) (몫, 나머지)

print("\n거듭제곱 (pow()):")
print(pow(x, y))  # 1000

print("\n거듭제곱 (**):")
print(x ** y)  # 1000

# 추가 예제
print("\n복합 연산:")
z = 2
print((x + y) * z)  # 26

print("\n부동소수점 연산:")
a = 0.1
b = 0.2
print(a + b)  # 0.30000000000000004 (부동소수점 정밀도 이슈)

print("\n문자열 반복:")
print("Hello " * 3)  # Hello Hello Hello

print("\n리스트 연산:")
list1 = [1, 2]
list2 = [3, 4]
print(list1 + list2)  # [1, 2, 3, 4]
print(list1 * 2)  # [1, 2, 1, 2]
```
