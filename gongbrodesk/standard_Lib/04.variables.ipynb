## 변수(Variables)

### 파이썬의 변수:

#### 동적 타이핑: 파이썬은 변수의 타입을 자동으로 결정합니다.
#### 참조: 변수는 객체에 대한 참조입니다.
#### 네이밍 규칙: 문자, 숫자, 밑줄(_)을 사용하며, 숫자로 시작할 수 없습니다.
#### 대소문자 구분: 'name'과 'Name'은 다른 변수입니다.
#### 예약어: 'if', 'for' 등의 예약어는 변수명으로 사용할 수 없습니다.

##### 변수명은 이름만 봐도 알 수 있도록 지어줘야 한다.

```python
# 기본 변수 할당
x = 5
y = "Hello"
print(f"x = {x}, y = {y}")

# 동적 타이핑
x = "Now I'm a string"
print(f"x is now: {x}")

# 여러 변수 동시 할당
a, b, c = 1, 2, 3
print(f"a = {a}, b = {b}, c = {c}")

# 변수 타입 확인
name = "Alice"
age = 30
is_student = False
print(f"Type of name: {type(name)}")
print(f"Type of age: {type(age)}")
print(f"Type of is_student: {type(is_student)}")

# 변수 참조
list1 = [1, 2, 3]
list2 = list1
list2.append(4)
print(f"list1: {list1}")  # list1도 변경됨

# 변수 삭제
del x
# print(x)  # 이 줄의 주석을 제거하면 NameError 발생

# 전역 변수와 지역 변수
global_var = "I'm global"

def test_function():
    local_var = "I'm local"
    print(global_var)  # 전역 변수 접근 가능
    print(local_var)

test_function()
print(global_var)
# print(local_var)  # 이 줄의 주석을 제거하면 NameError 발생

# 상수 (파이썬에는 진정한 상수는 없지만, 관례상 대문자로 표기)
PI = 3.14159
MAX_SPEED = 120

print(f"PI = {PI}, MAX_SPEED = {MAX_SPEED}")
```

### string 변수클래스의 메서드들.
  
  #### upper(), lower()
  #### strip(), lstrip() , rstrip()
  #### split()
  #### join()
  #### replace()
  #### startwith(), endwith()
  #### find(), index()
  #### count()
  #### isalhpa(), isdigit(), isalnum()
  #### format()
  

```python


# upper(), lower()

# 문자열을 대문자 또는 소문자로 변환합니다.

pythonCopys = "Hello, World!"
print(pythonCopys.upper())  # HELLO, WORLD!
print(pythonCopys.lower())  # hello, world!

# strip(), lstrip(), rstrip()

# 문자열의 앞뒤(또는 왼쪽, 오른쪽) 공백을 제거합니다.

pythonCopys = "  hello  "
print(pythonCopys.strip())  # "hello"
print(pythonCopys.lstrip())  # "hello  "
print(pythonCopys.rstrip())  # "  hello"

# split()

# 문자열을 특정 구분자로 나눕니다.

pythonCopys = "apple,banana,orange"
print(pythonCopys.split(','))  # ['apple', 'banana', 'orange']

# join()

# 문자열 리스트를 하나의 문자열로 합칩니다.

pythonCopyfruits = ['apple', 'banana', 'orange']
print(','.join(pythonCopyfruits))  # "apple,banana,orange"

# replace()

# 문자열 내의 특정 부분을 다른 문자열로 대체합니다.

pythonCopys = "Hello, World!"
print(pythonCopys.replace('World', 'Python'))  # "Hello, Python!"

# startswith(), endswith()

# 문자열이 특정 부분으로 시작하거나 끝나는지 확인합니다.

pythonCopys = "Hello, World!"
print(pythonCopys.startswith('Hello'))  # True
print(pythonCopys.endswith('World!'))  # True

# find(), index()

# 문자열 내에서 특정 부분 문자열의 위치를 찾습니다.

pythonCopys = "Hello, World!"
print(pythonCopys.find('World'))  # 7
print(pythonCopys.index('World'))  # 7
# find는 찾지 못하면 -1 반환, index는 예외 발생

# count()

# 문자열 내에서 특정 부분 문자열의 등장 횟수를 셉니다.

pythonCopys = "Hello, Hello, World!"
print(pythonCopys.count('Hello'))  # 2

# isalpha(), isdigit(), isalnum()

# 문자열이 알파벳, 숫자, 또는 알파벳+숫자로만 이루어졌는지 확인합니다.


print("123".isdigit())  # True
print("abc123".isalnum())  # True


# format()
# 문자열 포맷팅을 수행합니다.

pythonCopyname = "Alice"
age = 30
print("My name is {} and I'm {} years old".format(pythonCopyname, age))
# "My name is Alice and I'm 30 years old"



# capitalize()

# 문자열의 첫 글자를 대문자로 변환하고, 나머지 글자는 소문자로 변환합니다.
# 문자열의 첫 글자가 이미 대문자인 경우에는 변화가 없습니다.
# 문자열의 첫 글자가 알파벳이 아닌 경우, 그 글자는 그대로 유지되고 나머지 글자만 소문자로 변환됩니다.

# 기본적인 사용
s1 = "hello world"
print(s1.capitalize())  # "Hello world"

# 이미 첫 글자가 대문자인 경우
s2 = "Hello World"
print(s2.capitalize())  # "Hello world"

# 첫 글자가 알파벳이 아닌 경우
s3 = "123 hello world"
print(s3.capitalize())  # "123 hello world"

# 모든 글자가 대문자인 경우
s4 = "HELLO WORLD"
print(s4.capitalize())  # "Hello world"

# 공백으로 시작하는 경우
s5 = " hello world"
print(s5.capitalize())  # " hello world"
```
