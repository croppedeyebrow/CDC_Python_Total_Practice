# 파이썬 원시 자료형(Primitive Data Types) 심화 정리

- AI/ML/DL 엔지니어를 위한 CPython 내부 구조 관점 -

## 0. 왜 "원시 자료형"부터 다시 봐야 하는가?
=> 파이썬에는 엄밀한 의미의 "primitive type"이 없다.
C/Java와는 달리, int/float/bool 조차 전부 "객체(object)"이다.
   --> 이 사실 하나가 AI/ML 코드의 성능,메모리,정밀도 문제의 근원이 됨.
   --> NumPy/PyTorch가 존재하는 이우도 결국
   파이썬 스칼라는 느리고 무겁다라는 문제를 우회하기 위함.

## 1. 전체 타입 계층 개요.


| 타입 | 설명/분류 | 가변성 |예시| C 레벨 구조체 |
|---|---|---|---|
| `int` | 정수 | Immutable | 5, -100, 1000000000000000000 |`PyLongObject` |
| `float` | 실수 (배정밀도) | Immutable | 3.14, -0.001, 1.0e-10 | `PyFloatObject` |
| `complex` | 복소수(거의 안씀, 신호처리 등 특수 목적용) | Immutable | 1+2j, -3-4j | `PyComplexObject` |
| `bool` | 불리언 (int의 서브클래스) | Immutable | True, False | `PyLongObject` 재사용 |
| `str` | 문자열 | Immutable | "Hello, World!", "안녕하세요" | `PyUnicodeObject` |
| `bytes` | 바이트열 | Immutable | b"Hello, World!", b"안녕하세요" | `PyBytesObject` |
| `NoneType` | 널 값 | Immutable (싱글턴) | None |`_Py_NoneStruct` |
| `list`, `tuple`, `dict`, `set` | 컨테이너 (본 문서 범위 외) | 가변/불변 혼재 | — |


* 핵심 통찰 : 파이썬의 "원시"값은 전부 '힙'에 할당된 객체이며, 변수는 그 객체를 가리키는 참조(포인터)일뿐.
즉 a =5 는 "5라는 값을 a에 저장"이 아니라,
"정수 객체 5를 참조하는 포인터를 a라는 이름에 바인딩"하는 것.

- a =5 a는 정수 객체 5를 가리킴
- a= "hello" a는 문자열 객체 "hello"를 가리킴

## 2. int - 정수
:: 기본적으로 우리가 아는 정수 그대로.
+ 특이한 점 하나 : 파이썬은 정수 크기 제한이 없음.

- a = 10
  b = 2**100  처럼 어마어마하게 큰 수도 그냥 계산됨.

- C, Java 같은 언어는 int가 보통 32/64비트로 고정돼서 너무 큰 값을 넣으면 오버플로우가 발생.
   - 파이썬은 필요하면 알아서 크기를 늘려서 저장.
   - 편하지만 대신 크기가 커질수록 계산이 느려짐.

   ### 알아두면 좋은 것 - 작은 정수는 재사용된다.  

## 3. float - 실수

:: 소수점 있는 수.
- 내부적으로는 컴퓨터가 2진수로 저장하다 보니 딱 떨어지지 않는 소수는 근사값으로 저장.

## 4. bool - 참/거짓.
:: 참/거짓 값을 표현하는 타입.
- 내부적으로는 int의 서브클래스이며, 0이면 False, 1은 True로 저장.

## 5. str - 문자열
- 텍스트를 다루는 타입
- 파이썬3부터는 문자열이 전부 유니코드라서 한글,이모지,한자 등 모든 언어를 특별한 처리 없이 그냥 사용 가능.

## 6. NoneType - 널 값
- 값이 없다를 나타내는 특수한 값
- 다른 언어의 null, nill과 같은 개념.
- None은 프로그램 전체에서 딱 하나만 존재하는 값이라서,
   비교할 때 관례적으로 ==None이 아니라 is None으로 비교.


## 7. complex - 복소수
- 거의 안씀. 신호처리 등 특수 목적용.
- 예시 : 1+2j, -3-4j처럼 허수부를 가지는 수.
