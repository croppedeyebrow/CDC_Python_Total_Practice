# Numpy
- C언어 기반의 내부반복문을 사용한 속도가 매우 빠른 라이브러리.
- 다차원배열(ndarray)을 편리하게 사용하게 해주는 파이썬 라이브러리
- python은 list만 지원할 뿐, array 기능 자체가 없음.
   - EX) a=[1,2,3,4,5] b=['호랑이','사자','토끼','강아지','고양이']
- numpy를 가져와서 다차원배열에 대한 논리적계산, 정렬, 통계 ,시물레이션 등을 할 수 있음.


- 리스트는 붙이고 떼고 하는 것은 가능하지만, 리스트 끼리 계산하는 것은 불가능.

- 대량의 데이터를 array형태로 가지고 와서 고급진 수학, 과학 연산을 수행함.
     - 데이터 사이언스, 데이터분석, 딥러닝, 머신러닝을 구현하는 기본적인 라이브러리.
- <import numpy as np> -> numpy를 라이브러리로 호출해서 np로 부르기로 함.



# List vs Array

- array : 적은 메모리로 데이터를 빠르게 처리하고 모든 원소는 자료형을 가짐.
- list : 상대적으로 속도가 매우 느리고 원소가 각각 다른 자료형을 가질 수 있음.



# ndarray

- N-dimensional array(다차원 배열)
- 데이터 타입은 모두 동일함.
- 내부반복문을 사용함으로써 속도가 빠름.
- 배열간에 산술 연산이 가능함.

# 데이터 타입의 종류.

- int(8bit,16bit,64bit) i1,i2,i4,i8
- unit(8bit,16bit,32bit,64bit) u1,u2,u4,u8
- float(16bit,32bit,64bit,128bit) f2,f4,f8,f16
- 복소수(complex) 형
     - complex64: 두개의 32bit, 부동소수점 복소수 c8
     - complex128 : 두개의 64bit, 부동소수점 복소수 c16
- unicode : 고정길이 문자열
- bool : True, False

# dtype / astype/ arange / reshape

   - dtype : 자료형 확인
   - astype : 자료형 변환
   - np.arrange(start,stop,step) : step에 따라 증가하는 수열생성
   - reshape: 만들어진 배열의 데이터를 그대로 가지고 형태를 바꿈.
