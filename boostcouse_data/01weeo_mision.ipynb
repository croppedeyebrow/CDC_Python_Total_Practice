# 1번 문제
여러분은 파이썬을 통해 설문조사 문항의 응답내역을 분석하게 되었습니다. 문항별 응답내용에는 하나의 응답만 할 수 있는 single choice 문제와 여러 응답을 선택할 수 있는 multiple choice 문제가 있습니다. 2개를 구분하기 위해 single choice 문항 번호에 "_"를 표기하지 않기로 했습니다.

문항별 응답내역이 'question'에 담겨 있을 때, 조건문과 반복문을 사용하여 아래와 같은 결과가 출력되도록 코드를 작성해보세요.

````python
def get_single_choice_questions(questions):
    """
    주어진 질문 리스트에서 single choice 문항만 추출하는 함수
    
    :param questions: 전체 질문 리스트
    :return: single choice 문항만 포함된 리스트
    """
    return [q for q in questions if '_' not in q]

# 응답 설문 문항
question = ['Q2', 'Q3', 'Q4', 'Q5', 'Q6_1', 'Q6_2', 'Q6_3', 'Q6_4', 'Q6_5', 'Q6_6',
       'Q6_7', 'Q6_8', 'Q6_9', 'Q6_10', 'Q6_11', 'Q6_12', 'Q7_1', 'Q7_2',
       'Q7_3', 'Q7_4', 'Q7_5', 'Q7_6', 'Q7_7', 'Q8', 'Q9', 'Q10_1', 'Q10_2',
       'Q10_3']

# 함수 호출 및 결과 출력
single_choice_questions = get_single_choice_questions(question)
print(single_choice_questions)
````

## 2번문제
한스 로슬링(Hans Rosling, 1948년 7월 27일 ~ 2017년 2월 7일)은 스웨덴의 의사이자 통계학자로 비영리 벤처 갭마인더 재단의 공동설립자이기도 합니다. 빅데이터를 가장 잘 활용하는 보건 통계학자로 알려져 있습니다. 베스트셀러 책인 "팩트풀니스" 저자이기도 합니다. 갭마인더 사이트에서는 연도별, 국가별 GDP와 기대수명 데이터를 제공하고 있는데, 대표적으로 파이썬 라이브러리 중 'seaborn'에서 제공되는 예제 데이터가 있습니다. 오늘은 이 데이터를 활용해 문제를 풀어보겠습니다.
다음의 데이터는 연도, 국가별 기대수명을 나타내고 있는 데이터 입니다. 2011년 부터의 연도별, 국가별 평균 기대수명을 구해주세요. (2011년도 포함되게 구합니다.)

[필수 조건] groupby 나 pivot_table을 활용합니다. groupby 로 구할 때 unstack()이라는 기능을 사용하여 아래와 같이 컬럼에 인덱스 값을 올려서 표기할 수 있습니다.
만약 행에는 '연도'가, 열에 '국가'가 들어있고 수치 데이터의 결과값이 아래와 같다면, 출력형태는 조금 달라도 괜찮습니다.
데이터는 다음의 방법으로 읽어옵니다.
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df

````python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")

# 데이터 프레임의 열 이름 확인
print(df.columns)

# 2011년 이후 데이터만 필터링
df_filtered = df[df['Year'] >= 2011]

# 연도별, 국가별 평균 기대수명 계산
# groupby 함수:
#   - 'Year'와 'Country'를 기준으로 데이터를 그룹화합니다.
#   - 각 그룹 내에서 'Life_Expectancy' 열의 평균을 계산합니다.
# unstack 함수:
#   - 다중 인덱스 결과를 2차원 테이블 형태로 변환합니다.
#   - 'Country'가 열로 이동하고, 'Year'는 행 인덱스로 남습니다.
result = df_filtered.groupby(['Year', 'Country'])['Life_Expectancy'].mean().unstack()

# 결과 출력
print(result[['Canada', 'France', 'Germany', 'Great Britain', 'Japan', 'USA']])
````

# Python Basics

## Import

파이썬에서는 외부 라이브러리나 모듈을 사용하기 위해 `import`를 사용합니다.

```python
import pandas as pd
import numpy as np


Boolean
True와 False는 파이썬의 불리언 값입니다.
True는 1과 같고, False는 0과 같습니다.
python
print(True)  # 결과: True
print(False)  # 결과: False

print(True == 1)  # 결과: True
print(False == 0)  # 결과: True

print(True and False)  # 결과: False
print(True or False)  # 결과: True



Data Type
파이썬의 주요 데이터 타입에는 정수(int), 부동 소수점(float), 문자열(str), 리스트(list) 등이 있습니다.
python
x = 10  # int
y = 3.14  # float
name = "Alice"  # str
numbers = [1, 2, 3, 4, 5]  # list

List
리스트는 여러 값을 하나의 변수에 저장할 수 있는 데이터 타입입니다.
python
# 리스트 생성
fruits = ["apple", "banana", "cherry"]

# 리스트 원소 접근
print(fruits)  # 결과: apple
print(fruits[-1])  # 결과: cherry

# 리스트에 원소 추가
fruits.append("date")
print(fruits)  # 결과: ['apple', 'banana', 'cherry', 'date']

For
for 루프는 반복문을 사용하여 리스트나 다른 시퀀스를 순회합니다.
python
# 리스트의 원소를 하나씩 출력
for fruit in fruits:
    print(fruit)
# 결과: apple
#         banana
#         cherry
#         date

If
if 문은 조건문을 사용하여 특정 조건이 참인지 여부에 따라 코드를 실행합니다.
python
x = 10
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
# 결과: x is greater than 5

Len
len() 함수는 문자열, 리스트 등의 길이를 반환합니다.
python
print(len(fruits))  # 결과: 4
print(len("Hello, world!"))  # 결과: 13

Split
split() 메서드는 문자열을 특정 구분자로 나누어 리스트로 반환합니다.
python
sentence = "Hello, world! Welcome to Python."
words = sentence.split()
print(words)  # 결과: ['Hello,', 'world!', 'Welcome', 'to', 'Python.']

Join
join() 메서드는 리스트의 문자열을 특정 구분자로 연결하여 하나의 문자열로 반환합니다.
python
words = ['Hello,', 'world!', 'Welcome', 'to', 'Python.']
sentence = " ".join(words)
print(sentence)  # 결과: 'Hello, world! Welcome to Python.'

# Pandas Basics

## Pandas

`pandas`는 데이터 조작 및 분석을 위한 파이썬 라이브러리입니다.

```python
import pandas as pd

DataFrame
DataFrame은 2차원 데이터 구조로, 행과 열로 구성됩니다.
python
# DataFrame 생성
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)
print(df)
# 결과:
#       Name  Age
# 0    Alice   25
# 1      Bob   30
# 2  Charlie   35

Series
Series는 1차원 데이터 구조로, DataFrame의 한 열을 나타냅니다.
python
# Series 생성
ages = pd.Series([25, 30, 35], name='Age')
print(ages)
# 결과:
# 0    25
# 1    30
# 2    35
# Name: Age, dtype: int64

Sort
sort_values() 메서드는 데이터를 정렬합니다.
python
# 나이 기준으로 정렬
df_sorted = df.sort_values(by='Age')
print(df_sorted)
# 결과:
#       Name  Age
# 0    Alice   25
# 1      Bob   30
# 2  Charlie   35

Drop
drop() 메서드는 특정 행이나 열을 제거합니다.
python
# 'Age' 열 제거
df_dropped = df.drop(columns=['Age'])
print(df_dropped)
# 결과:
#       Name
# 0    Alice
# 1      Bob
# 2  Charlie

# 첫 번째 행 제거
df_dropped_row = df.drop(index=0)
print(df_dropped_row)
# 결과:
#       Name  Age
# 1      Bob   30
# 2  Charlie   35

GroupBy
groupby() 메서드는 데이터를 특정 기준으로 그룹화합니다.
python
# 'Age' 기준으로 그룹화하고 평균 계산
grouped = df.groupby('Age').mean()
print(grouped)
# 결과:
#       Age
# Age      
# 25      25
# 30      30
# 35      35

Pivot Table
pivot_table() 메서드는 데이터를 피벗 테이블 형식으로 변환합니다.
python
# 피벗 테이블 생성
pivot = df.pivot_table(values='Age', index='Name', aggfunc='mean')
print(pivot)
# 결과:
#          Age
# Name        
# Alice    25
# Bob      30
# Charlie  35

Plot
plot() 메서드는 데이터를 시각화합니다. pandas는 matplotlib와 통합되어 있습니다.
python
import matplotlib.pyplot as plt

# 나이 분포를 막대 그래프로 시각화

plt.show()
